# Generating Random Numbers
* Random numbers are often required to randomly sample data, generating synthetic distribution of samples, generating noise etc.
* Random numbers can have uniform distribution, normal distribution etc.
* Here we shall first learn to generate uniform and normal random numbers using numpy, check their histogram then
    * Use random number for generating some synthetic X-Y data
    * X shall be in range -10 to 10
    * Y shall depend on X in some functional form + some random noise

```python
# Import basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generating random(uniform) numbers using numpy
uni_r=np.random.uniform(-15,15, 10000)    # Means, generate 10000 random numbers within range -15 to 15 uniformly  
uni_r
```

```python
# Histogram plot of the random numbers
plt.hist(uni_r, bins=50)
```

```python
# Generating random(normal) numbers using numpy
nor_r=np.random.normal(0,1, 10000)    # Means, generate 10000 random numbers centred around 0(called mean) and having standard deviation=1  
nor_r
```

```python
# Histogram plot of the random numbers
plt.hist(nor_r, bins=50)
```

# Generating Synthetic Linear data + Random Noise

```python
# Linspace function ias used to create evenly spaced numbers over a specified interval(here interval is -10 to 10 and number of samples=101
x=np.linspace(-10, 10, 101)
x
```

```python
# Using numpys array's property of doing mathematical calculations on array
y0=12*x+30   # Using linear relation
y0
```

```python
# Visualizing the data
plt.scatter(x,y0, s=4)
```

```python
# Length of array
size=len(y0)
size
```

```python
# Generate 101 number of random numbers(noise)
np.random.seed(42)
y_noise=np.random.uniform(-12,12, size)
# Add the generated noise to the synthetic data
y=y0+y_noise
y
```

```python
# Visualize the data with noise
plt.scatter(x,y)
```

```python
# Generate a panda dataframe to store the X, Y data
df=pd.DataFrame()
df['X']=x
df['Y']=y
df
```

```python
# Save the data to csv for using later
df.to_csv('Linear_data.csv')
```

# Generating Synthetic Non-Linear data + Random Noise

```python
y_noise=np.random.uniform(-20,20, size)
y=4+10*x+2*x**2-0.5*x**3+y_noise        # Using some non-linear(polynomial) function
```

```python
plt.scatter(x,y)
```

```python
df1=pd.DataFrame()
df1['X']=x
df1['Y']=y
df1
df1.to_csv('NonLinear_data.csv')
```
